### 1. 請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。




```python
# 記得先 Import 正確的套件
import numpy as np
import timeit
```

```python
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
```

```python
%timeit -n 10 a = 2
Z=Z+1
```

```python
%timeit -n 10 a = 2
for matrix in Z:
    for row in matrix:
        for element in row:
            element+1
```

```python
%timeit -n 10 a = 2
for elements in Z.flat:
    elements+1
```

```python
%timeit -n 10 a = 2
for d in np.nditer(Z):
    d+1
```

### 2. 如何從一個陣列中，找出出現頻率最高的數值與位置？


```python

Z = np.random.randint(0,10,50)
print(Z)
```

```python
unique , counts = np.unique(Z, return_counts = True)
a = unique[np.argmax(counts)]
print("Highest frequency number =", a)
print("Position =", np.where(Z==a))
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
```

```python
b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b)) #It's not a good idea
newListB=[]
for row in b:
    newListB.append(list(row))
print(newListB)
```

```python
c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c)) #It's not a good idea
newListC=[]
for matrix in c:
    for row in matrix:
        newListC.append(list(row))
print(newListC)
```
