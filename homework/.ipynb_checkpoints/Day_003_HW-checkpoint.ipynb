### 1. [簡答題] 請比較 np.zeros 和 np.empty 產生出來的陣列有何差異？為什麼要設計兩種方法？

```
print(np.zeros((2, 3)))
print(np.empty((2, 3)))
```


````python
import numpy as np
print(np.empty((2, 3)))
print(np.zeros((2, 3)))
print("zeros()產出全為0的array\nempty()產出未初始化的陣列")
````

### 2. 在不用「整數亂數方法」的限制下，如何將包含小數的轉換整數？請將給定的 a 陣列當中的元素變成去掉小數變成整數。


````python
# 記得先 Import 正確的套件
import numpy as np

a = np.random.rand(2,3)
print(a)
print("use around() method")
print(np.around(a))
````

### 3. 承上題，怎樣可以限制整數的範圍介於 m - n 之間？請將給定的 a 陣列當中的元素的範圍調整成 m - n 之間。


````python
m = 20
n = 40
a = np.random.rand(2,3)*(n-m) #生成0~20之random矩陣
print(a)
b = np.around(a) #對a四捨五入
print(b+m) #再加20回來, 得到20~40的範圍之矩陣
````
