### 1. [簡答題] 請問 type(...) 跟 a.dtype 這兩個語法有什麼不同？

```
print('type(a): ', type(a))
print('a.dtype: ', a.dtype)
```


````python
import numpy as np
a=np.array(range(51))
print('type(a): ', type(a))
print('a.dtype: ', a.dtype)

'''
上述code得知
type(a)表示 "a" 這個物件的型別
而 a.dtype則是 "a" 裡面包的資料型態 (int64)
'''
````

### 2. 請撰寫一個判斷 a 的元素是否等於指定資料型態的函式，並且通過測試。



````python
# 記得先 Import 正確的套件
import numpy as np
````

````python
def is_dtype(a, t):
    return a.dtype==t
````

````python
a = np.random.randint(10, size=6) 

# 測試一
print(is_dtype(a, 'int')) # True

# 測試二
print(is_dtype(a, np.int)) # True

# 測試三
print(is_dtype(a, np.dtype('int'))) # True
````

### 3. [簡答題] 承上題，請判斷下列三種寫法為何不正確？

```
def is_dtype(a, t):
    return a.dtype is t

def is_dtype(a, t):
    return type(a) == np.dtype(t)

def is_dtype(a, t):
    return type(a) is np.dtype(t)
```

````python
def is_dtype(a, t):
    return a.dtype is t
'''
錯誤
因為 "is" 是強比較
必須整個物件型別都相同
於是只會過test3
'''
````

````python
def is_dtype(a, t):
    return type(a) == np.dtype(t)
'''
錯誤
因為 "==" 
左邊的是整個class <class 'numpy.ndarray'>
右邊的是物件型別
'''
````
