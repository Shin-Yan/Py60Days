### 1. [簡答題] 請問下列這三種方法有什麼不同？

```
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
```



````python
import numpy as np
a = np.arange(12).reshape(3,4)
print(a.sum())
print(np.sum(a))
print(sum(a))
print("我們可以看到, sum(a)回傳的是對於矩陣每行的和")
print("\n而a.sum()和np.sum(a)的差異則較難直接看出來, \na.sum()是直接使用a這個np class 裡面的method, \nnp.sum(a)則是把a傳入np.sum這個函式")
````

### 2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
# 記得先 Import 正確的套件
import numpy as np
````

````python
A = np.random.random((5, 5))
print(A)
A=(A-A.mean())/(A.std())
print(A)
print(A.mean())
print(A.std())
````

### 3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
A = np.random.random(10)
A[np.argmax(A)]=-1
print(A)
````
